# Loading data

```python
iris = pd.read_csv('../data/iris.data', 
                           header=None, 
                           names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
```

# Exploratory data analysis

```python
iris.head()
```

```python
print('Iris dataset has {} rows and {} columns'.format(iris.shape[0], iris.shape[1]))
```

```python
X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = iris['class']
```

```python
colors = y.astype('category').cat.codes
pd.plotting.scatter_matrix(X, c=colors, figsize=[10, 10], s=150, marker='D')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X, y)
```

```python

X_novos = [[5.6, 2.8, 3.9, 1.1], 
           [5.7, 2.6, 3.8, 1.3], 
           [4.7, 3.2, 1.3, 0.2]]
predicao = knn.predict(X_novos)
print('Predicao: {}'.format(predicao))
```

```python
import joblib
filename = '../output/model/finalized_model.sav'
joblib.dump(knn, filename)
```
